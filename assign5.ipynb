```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("Mall_Customers.csv")
```

```python
df.head()
```

```python
x = df.iloc[:,3:]
```

```python
x
```

```python
plt.title("hh")
plt.xlabel("dj")
plt.ylabel("sh")
plt.scatter(x['Annual Income (k$)'] , x['Spending Score (1-100)'])
```

```python
from sklearn.cluster import KMeans , AgglomerativeClustering
```

```python
km= KMeans(n_clusters=4)
```

```python
km.fit_predict(x)
```

```python
km.inertia_  #SSE sum squared error 
```

```python
sse = []
for k in range(1,16) :
    km = KMeans(n_clusters=k)
    km.fit_predict(x)
    sse.append(km.inertia_)
```

```python
plt.title("Elbow Method")
plt.xlabel("Value of K")
plt.ylabel("SSE")
plt.grid()
plt.xticks(range(1,16))
plt.plot(range(1,16) , sse , marker=".")
```

```python
from sklearn.metrics import silhouette_score
```

```python
silh = []
for k in range (2,16):
 km = KMeans(n_clusters=k)
 labels= km.fit_predict(x)
 score  = silhouette_score(x,labels)
 silh.append(score)
```

```python
silh
```

```python
plt.title("Silhouette Method")
plt.xlabel("Value of K")
plt.ylabel("Silhoutee score")
plt.grid()
plt.xticks(range(2,16))
plt.bar(range(2,16) , silh)
```

```python
km = KMeans(n_clusters=5 , random_state=0)
labels = km.fit_predict(x)
```

```python
labels
```

```python
plt.title("Clustered Data")
plt.xlabel("Annual income")
plt.ylabel("spending score ")
plt.scatter(x['Annual Income (k$)'] , x['Spending Score (1-100)'] , c=labels)

cent = km.cluster_centers_
plt.scatter(cent[:,0],cent[:,1] , s=70 , color="r")
```

```python
agl = AgglomerativeClustering(n_clusters=5)
```

```python
alabels = agl.fit_predict(x)
```

```python
alabels
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title("K Means ")
plt.xlabel("Annual income")
plt.ylabel("spending score ")
plt.scatter(x['Annual Income (k$)'] , x['Spending Score (1-100)'] , c=labels)

cent = km.cluster_centers_
plt.scatter(cent[:,0],cent[:,1] , s=70 , color="r")

plt.figure(figsize=(16,9))
plt.subplot(1,2,2)
plt.title("Agglomerative Clustering ")
plt.xlabel("Annual income")
plt.ylabel("spending score ")
plt.scatter(x['Annual Income (k$)'] , x['Spending Score (1-100)'] , c=alabels)

cent = km.cluster_centers_
plt.scatter(cent[:,0],cent[:,1] , s=70 , color="r")

```
