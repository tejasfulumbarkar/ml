```python
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression 
import numpy as np
```

```python
df = pd.read_csv("temperatures.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
x= df['YEAR']
y = df['ANNUAL']
```

```python
x.shape
```

```python
plt.title("Temperatures of INDIA ")
plt.xlabel("year")
plt.ylabel("annual")
plt.scatter(x,y)
```

```python
x = x.values
x = x.reshape(117,1)
```

```python
x.shape
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x,y)
```

```python
regressor.coef_  # slope y = mx+c coef is m 
```

```python
regressor.intercept_  #c value in y = mx+c
```

```python
regressor.predict([[2046]])
```

```python
predicted = regressor.predict(x)
```

```python
predicted
```

```python
# mean absolute error 
np.mean(abs(y-predicted))
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
mean_absolute_error(y,predicted)
```

```python
mean_squared_error(y,predicted)
```

```python
from sklearn.metrics import r2_score   # r square metrics 
```

```python
r2_score(y,predicted)
```

```python
plt.title("Temperatures of INDIA ")
plt.xlabel("year")
plt.ylabel("annual")
plt.scatter(x,y , label="actual" , color='r')
plt.plot(x,predicted,label="predicted" , color='g')
plt.legend()
```
